```python
import pandas as pd

def load_data(file_path):
    ds = pd.read_csv(file_path)
    x_data = ds.iloc[:, :-1]
    y_data = ds.iloc[:, -1]
    return x_data, y_data
```

```python
def scale_data(x_data, scaler):
    return scaler.fit_transform(x_data)
```

```python
def train_and_predict(classifier, x_train, y_train, x_test):
    classifier.fit(x_train, y_train)
    return classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

def calculate_accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred) * 100
```

```python
import matplotlib.pyplot as plt

def plot_classifier_accuracies(nb_accuracies, svm_accuracies, lr_accuracies):
    labels = ['StandardScaler', 'MinMaxScaler', 'RobustScaler']
    bar_width = 0.25
    index = range(len(labels))

    plt.figure(figsize=(10, 6))

    plt.bar(index, nb_accuracies, bar_width, label='Naive Bayes')
    plt.bar([i + bar_width for i in index], svm_accuracies, bar_width, label='SVM')
    plt.bar([i + 2 * bar_width for i in index], lr_accuracies, bar_width, label='Logistic Regression')

    plt.xlabel('Scaler')
    plt.ylabel('Accuracy (%)')
    plt.title('Classifier Accuracies with Different Scalers')
    plt.xticks([i + bar_width for i in index], labels)
    plt.legend()
    plt.show()
```

```python
import pandas as pd

def create_accuracies_dataframe(labels, nb_accuracies, svm_accuracies, lr_accuracies):
    data = {'Scaler': labels, 'Naive Bayes Accuracy': nb_accuracies, 'SVM Accuracy': svm_accuracies, 'Logistic Regression Accuracy': lr_accuracies}
    accuracies_ds = pd.DataFrame(data)
    return accuracies_ds
```

```python
file_path = 'haberman.csv'
x_data, y_data = load_data(file_path)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scalers = [StandardScaler(), MinMaxScaler(), RobustScaler()]
accuracies_nb = []
accuracies_svm = []
accuracies_lr = []
```

```python
from sklearn.model_selection import train_test_split

def scale_train_plot(x_data, y_data, scaler, classifier, label):
    x_scaled = scale_data(x_data, scaler)
    x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_data, test_size=0.2, random_state=1)

    y_pred = train_and_predict(classifier, x_train, y_train, x_test)
    accuracy = calculate_accuracy(y_test, y_pred)
    return accuracy
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import  GaussianNB
from sklearn.svm import SVC

scalers = [StandardScaler(), MinMaxScaler(), RobustScaler()]

nb_classifier = GaussianNB()
svm_classifier = SVC(kernel='rbf')
lr_classifier = LogisticRegression()

nb_accuracies = []
svm_accuracies = []
lr_accuracies = []
```

```python
for scaler in scalers:
 nb_accuracy = scale_train_plot(x_data, y_data, scaler, nb_classifier, 'Naive Bayes')
 nb_accuracies.append(nb_accuracy)
```

```python
for scaler in scalers:
 svm_accuracy = scale_train_plot(x_data, y_data, scaler, svm_classifier, 'SVM')
 svm_accuracies.append(svm_accuracy)
```

```python
for scaler in scalers:
 lr_accuracy = scale_train_plot(x_data, y_data, scaler, svm_classifier, 'Logistic Regression')
 lr_accuracies.append(lr_accuracy)
```

```python
plot_classifier_accuracies(nb_accuracies, svm_accuracies, lr_accuracies)
```

```python
labels = ['StandardScaler', 'MinMaxScaler', 'RobustScaler']

accuracies_ds = create_accuracies_dataframe(labels, nb_accuracies, svm_accuracies, lr_accuracies)

print("Accuracies DataFrame:")
print(accuracies_ds)
```
